```python
import mlflow

import os

import dotenv
import pandas as pd

# Get the last run with an article_summarizer artifact
from mlflow.tracking import MlflowClient

client = MlflowClient()
runs = client.search_runs(
    #    experiment_ids=[client.get_experiment_by_name("Test Summarizer").experiment_id],
    experiment_ids=["0"],
    filter_string="tags.mlflow.runName LIKE '%'",
    order_by=["attribute.start_time DESC"],
)

print(runs)
```

```python
MODEL_NAME = "article_summarizer"
last_run = None
for run in runs:
    artifacts = client.list_artifacts(run.info.run_id)
    artifact_paths = [artifact.path for artifact in artifacts]
    if MODEL_NAME in artifact_paths:
        last_run = run
        break

print('Last run ID:', last_run.info.run_id)
print('Model path:', f"runs:/{last_run.info.run_id}/{MODEL_NAME}")
```

```python
# Load the model
model = mlflow.pyfunc.load_model(f"runs:/{last_run.info.run_id}/{MODEL_NAME}")

# Predict
text = open("assets/articles_short/ai_relationships.html").read()
input_example = pd.DataFrame({"text": [text]})
summary = model.predict(input_example)[0]

print(summary)
```
